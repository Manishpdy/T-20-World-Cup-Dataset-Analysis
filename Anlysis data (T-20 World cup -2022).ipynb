```python
def plot_batsman_performance(df):
    batsman_runs = df.groupby('wkt_batsman_name')['wkt_batsman_runs'].sum().sort_values(ascending=False)

    # Plot
    plt.figure(figsize=(12, 6))
    batsman_runs.nlargest(10).plot(kind='bar', color='skyblue')
    plt.title('Total Runs Scored by Batsmen')
    plt.xlabel('Batsman')
    plt.ylabel('Total Runs')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

plot_batsman_performance(df)
```

# Identify and analyze critical match events, such as boundaries, wickets, and strategic moments. Uncover the factors that contribute to the success or downfall of a team.

```python
boundaries = df[df['shortText'].str.contains('4 runs|6 runs')]
wickets = df[df['wicket_id'] >= 0]

# Analyze the distribution of boundaries and wickets
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

boundaries['over'].hist(bins=20, ax=axes[0], color='skyblue', alpha=0.7)
axes[0].set_title('Distribution of Boundaries Over Different Overs')
axes[0].set_xlabel('Over')
axes[0].set_ylabel('Frequency')

wickets['over'].hist(bins=20, ax=axes[1], color='salmon', alpha=0.7)
axes[1].set_title('Distribution of Wickets Over Different Overs')
axes[1].set_xlabel('Over')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
total_boundaries = boundaries.groupby('match_id').size().reset_index(name='total_boundaries')
total_wickets = wickets.groupby('match_id').size().reset_index(name='total_wickets')

match_outcomes = pd.merge(total_boundaries, total_wickets, on='match_id', how='inner')
match_outcomes['match_result'] = 'Win'
match_outcomes.loc[match_outcomes['total_wickets'] > match_outcomes['total_boundaries'], 'match_result'] = 'Loss'

result_counts = match_outcomes['match_result'].value_counts()

plt.figure(figsize=(8, 6))
result_counts.plot(kind='bar', color=['lightgreen', 'lightcoral'])
plt.title('Impact of Boundaries and Wickets on Match Outcomes')
plt.xlabel('Match Result')
plt.ylabel('Number of Matches')
plt.xticks(rotation=0)
plt.show()
```

# Evaluate individual and team performances, including standout players, consistent run-scorers, and top wicket-takers. Discover what sets them apart.

```python
def plot_standout_players(df):
    standout_players_runs = df.groupby('wkt_batsman_name')['wkt_batsman_runs'].sum().nlargest(5)
    other_players_runs = df['wkt_batsman_runs'].sum() - standout_players_runs.sum()

    # Combine standout players and others
    standout_players_runs['Others'] = other_players_runs

    # Plot
    plt.figure(figsize=(6, 6))
    plt.pie(standout_players_runs, labels=standout_players_runs.index, autopct='%1.1f%%', startangle=140)
    plt.title('Contribution of Standout Players to Total Runs')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
plot_standout_players(df)
```

```python
def plot_consistent_run_scorers(df):
    # Group runs scored by batsman name
    player_runs = df.groupby('wkt_bowler_name')['bowler2_wkts'].sum()

    # Calculate mean runs scored
    mean_runs = player_runs.mean()

    # Filter consistent run-scorers
    consistent_run_scorers = player_runs[player_runs >= mean_runs]

    # Calculate runs scored by other players
    other_players_runs = df['bowler2_wkts'].sum() - consistent_run_scorers.sum()

    # Add 'Others' category
    consistent_run_scorers['Others'] = other_players_runs

    # Plot
    plt.figure(figsize=(14, 8))
    consistent_run_scorers.plot(kind='bar', color='skyblue')
    plt.title('Distribution of Runs Scored by Consistent Run-Scorers')
    plt.xlabel('Runs')
    plt.ylabel('Batsman')
    plt.show()

# Assuming df is your DataFrame containing the dataset
plot_consistent_run_scorers(df)
```

```python

def plot_top_wicket_takers(df):
    # Group wickets taken by bowler name and sum across all matches
    top_wicket_takers = df.groupby('wkt_bowler_name')['bowler2_wkts'].sum().nlargest(5)

    # Calculate total wickets taken by all other bowlers
    other_bowlers_wickets = df['bowler2_wkts'].sum() - top_wicket_takers.sum()

    # Combine top wicket-takers and others
    top_wicket_takers['Others'] = other_bowlers_wickets

    # Plot
    plt.figure(figsize=(10, 6))
    top_wicket_takers.plot(kind='bar', color='skyblue')
    plt.title('Top Wicket-Takers')
    plt.xlabel('Wickets Taken')
    plt.ylabel('Bowler')
    plt.grid(True, axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Assuming df is your DataFrame containing the dataset
plot_top_wicket_takers(df)
```
